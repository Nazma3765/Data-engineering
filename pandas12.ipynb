```python
#22.Count transactions made on weekends.
df['dayofweek'] = df['invoice_date'].dt.dayofweek
weekend_transactions = df[df['dayofweek'].isin([5, 6])]
count_weekends = len(weekend_transactions)
print(f"Number of transactions on weekends: {count_weekends}")
```

```python
#23.Find which shopping mall had the highest sales in December.
dec_sales = df[df['invoice_date'].dt.month == 12].groupby('shopping_mall')['total_price'].sum()
top_mall = dec_sales.idxmax()   #idxmax() finds the mall name with the highest sales.
top_sales = dec_sales.max()     #max() gives the sales amount.
print(f"The shopping mall with the highest sales in December is {top_mall} with total sales of {top_sales}.")
```

```python
#26.Plot daily revenue trends.
daily_revenue = df.groupby(df['invoice_date'].dt.date)['total_price'].sum()
plt.figure(figsize=(10, 5))
daily_revenue.plot(kind='line', marker='o')
plt.title('Daily Revenue Trend')
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#27.#27.Determine which day of the week has the most transactions.
df['day_of_week'] = df['invoice_date'].dt.day_name()
day_counts = df['day_of_week'].value_counts()
most_transactions_day = day_counts.idxmax()
print(f"The day with the most transactions is: {most_transactions_day}")
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("C:/Users/Nazma/Downloads/customer_shopping_data.csv/customer_shopping_data.csv")

# Step 1: Convert invoice_date to datetime
df['invoice_date'] = pd.to_datetime(df['invoice_date'], errors='coerce')

# Step 2: Extract the month in 'YYYY-MM' format
df['month'] = df['invoice_date'].dt.to_period('M')

# Step 3: Group by month and shopping_mall, then sum the revenue
monthly_revenue = df.groupby(['month', 'shopping_mall'])['price'].sum().reset_index()

# Step 4: Show result
print(monthly_revenue.head())
```

```python
#Aggregate daily total quantity sold.
daily_quantity = df.groupby(df['invoice_date'].dt.date)['quantity'].sum()
daily_quantity.columns = ['date', 'total_quantity']
print(daily_quantity.head())
```

```python
#Find the month with the lowest average price.
lowest_avg_price=df.groupby('month')['price'].mean()
print(lowest_avg_price)
```
